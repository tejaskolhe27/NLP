## Q. 1

### 1. Read a word from user and print all the meaning of it

```python
word_sys = input("Enter a word to find the meaning : ")

for synset in wordnet.synsets(word_sys):
    
    
    if word_sys not in synset.name():
        
        print(synset.name(),synset.lemmas()[0].name(), synset.definition())
        
        print()
```

### 2. Read a word from user and print all noun meaning of it

```python
word_sys = input("Enter a word to find the meaning : ")

for synset in wordnet.synsets(word_sys):
    
    if '.n.' in synset.name():
        
        print(synset.name(),synset.lemmas()[0].name(), synset.definition())
        
        print()
```

### 1. Apply chunking and show the chunks of type <[VBD]><[DT]>

```python

''' 
    According to the rule you created,yours  chunks:
    
    Start with an optional (?) Verb past tense(<VBD>)
    Can have any number(*) determiner (DT)
'''

grammer = "VD:{<VBD><DT>}"
```

```python
# Initializing the regex parser

reg_parser = RegexpParser(grammar = grammer)

tags = pos_tag(word_tokenize(text))

tree = reg_parser.parse(tags)

tree
```

### 2. Find all the named entities present in article

```python
nlp = spacy.load('en_core_web_sm')
```

```python
sent = nlp(text = text)
```

```python
# iterating over the input data to recognize named entity

ner = {}

for ent in sent.ents:
    
    ner[ent.text] = ent.label_
    
displacy.render(sent, style = 'ent', jupyter = True)
```

### 3. Access all the verbs and apply the porter stemmer and lancaster stemmer

```python
# Converting all the words of the input text into tokens

word_token = word_tokenize(text)

tags = pos_tag(word_token)
```

#### a. Porter Stemmer

```python
# Initializing the porter stemmer

pr_stemmer = PorterStemmer()
```

```python
# Iterating over the tags to segregate verbs

verb_text = [word for word, tags in tags if tags.startswith('V')]

porter_stem = set()

for word in verb_text:
    
    word_stem = pr_stemmer.stem(word = word)
    
    porter_stem.add(word_stem)
    
    if word not in porter_stem:
        
        print('Stemmed form of ',word, 'is ', word_stem)
```

#### b. Lancaster Stemmer

```python
# Initalizing the Lancaster Stemmer

lncst_stemmer = LancasterStemmer()
```

```python
# Iterating over the word tags and applying Lancaster stemmer method

lan_stem = set()

for word in verb_text:
    
    word_lan_stem = lncst_stemmer.stem(word)
    
    lan_stem.add(word_lan_stem)
    
    if word not in lan_stem:
        
        print('Stemmed form of ',word, 'is ', word_lan_stem)
```

### 4. Access all the verbs in past tense and Lemmatize them

```python
# Initializing the wordnet Lemmatizer

wn_lemmatizer = WordNetLemmatizer()
```

```python
# Segregating all the verb in past tense

word_verb_past = [word for word,tag in tags if tag == 'VBD']

for word in word_verb_past:
    
    print('Lemmatized form of ', word, 'is ',wn_lemmatizer.lemmatize(word, pos = 'v'))
```

### 5. Read only the adjective from the text and find their synonyms

```python
# Segregating adjective

word_adjective = [word for word, tag in tags if tag == 'JJ']

# Finding synonyms 

synonym = {}

for word in  word_adjective:
    
    for synset in wordnet.synsets(word):
        
        synonym[word] = synset.lemmas()[0].name()
    
# Printing all the synonyms
    
for key, value in synonym.items():
    
    print('Synonym of ', key, 'is', value)
```

### 6. Read all the verbs and print their antonyms

```python
# Segreagating all the verbs

word_verb = [word for word, tag in tags if tag.startswith('V')]

antonym = {}

for word in word_verb:
    
    for synset in wordnet.synsets(word):
        
        for lemma in synset.lemmas():
            
            if lemma.antonyms():
                
                antonym[word] = lemma.antonyms()[0].name()
                
# Printing all the antonyms
    
for key, value in antonym.items():
    
    print('Antonym of ', key, 'is', value)
```

### 7. Print all the entities marked as DATE and TIME in the data

```python
# Iterating over the ner dictionary to retrieve all the entities marked as DATE and TIME

for key, value in ner.items():
    
    if value == 'DATE' or value == 'TIME':
        
        print(key, '---->', value)
```

### 8. Print all the named of location present in the article

```python
# Iterating over the ner dictionary to retrieve all the entities marked as LOC

for key, value in ner.items():
    
    if value == 'LOC':
        
        print(key, '---->', value)
```
