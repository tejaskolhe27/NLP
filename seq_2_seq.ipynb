````python
import pandas as pd
import numpy as np
import os

path = '/content/drive/MyDrive/Articles/'
````

````python
headlines = []
for filename in os.listdir(path):
  if "Articles" in filename:
    article_df = pd.read_csv(path + filename)
    headlines.extend(list(article_df['headline'].values))
````

````python
len(headlines)
````

````python
headlines
````

````python
headlines = [h for h in headlines if h != 'Unknown']
print("The number of headlines is ",len(headlines))
````

````python
import string
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter(action = 'ignore',category = FutureWarning)
````

````python
def clean_text(headline):
  text = "".join(word for word in headline if word not in string.punctuation ).lower()
  text = text.encode('utf8').decode("ascii",'ignore')
  return text

corpus = [clean_text(headline)for headline in headlines]
````

````python
corpus[:5]
````

````python
vocab = []

for line in corpus:
  words = line.split()
  for word in words:
    vocab.append(word)

vocabulary = set(vocab)
````

````python
len(vocabulary)
````

````python
from keras_preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.utils import to_categorical


tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word2index  = tokenizer.word_index
len(word2index)
````

````python
#word2index
````

````python
dictionary ={}
rev_dictionary = {}

for word ,index in word2index.items():
  dictionary[word] = index
  rev_dictionary[index] = word
````

````python
input_sequences = tokenizer.texts_to_sequences(corpus)
````

````python
input_sequences[:5]
````

````python
input_data =[]
target = []
for line in input_sequences:
  for i in range(1,len(line)-1):
    input_data.append(line[:i])
    target.append(line[i])
````

````python
input_data
````

````python
MAX_LENGTH = 0
for seq in input_data:
  if len(seq)> MAX_LENGTH:
    MAX_LENGTH = len(seq)
MAX_LENGTH
````

````python
input_data = pad_sequences(input_data,maxlen = MAX_LENGTH,padding = 'post')
````

````python
input_data[:10]
````

````python
target = to_categorical(target,num_classes=len(word2index))
````

````python
target
````

````python
target.shape
````

````python
rev_dictionary[1]
````

````python
from keras.models import Sequential
from keras.layers import LSTM,Dense,Embedding,Dropout,GRU
from keras.callbacks import EarlyStopping


model = Sequential()
model.add(Embedding(input_dim = len(vocabulary),output_dim = 100,input_length=MAX_LENGTH))


model.add(LSTM(100))
model.add(Dropout(rate = 0.1))


# model.add(LSTM(100))
# model.add(Dropout(0.2))


model.add(Dense(units = target.shape[1],activation='softmax'))
````

````python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
````

````python
model.fit(input_data,target,batch_size = 10,epochs = 10)
````
