### Removing Punctuations

```python
a = """
Sachin Ramesh Tendulkar, born 24 April 1973 is an Indian former international 
cricketer who captained the Indian national team. He is widely regarded as one of the greatest 
batsmen in the history of cricket. Hailed as the world's most prolific batsman of all time, 
he is the all-time highest run-scorer in both ODI and Test cricket with more than 18,000 runs 
and 15,000 runs, respectively. He also holds the record for receiving the most player of the 
match awards in international cricket.Tendulkar was a Member of Parliament, 
Rajya Sabha by nomination from 2012 to 2018.
"""
```

```python
import string
```

```python
s=string.punctuation
s
```

```python
from nltk import word_tokenize 
```

```python
token = word_tokenize(a)
```

```python
clean=[]
for i in token:
    if i not in s:
        clean.append(i)
clean=' '.join(clean)
clean
```

### Remove stopwords

```python
from nltk.corpus import stopwords
```

```python
swords=stopwords.words('english')
swords
```

```python
clean=clean.lower()
token1 = word_tokenize(clean)
```

```python
clean1=[]
for i in token1:
    if i not in swords:
        clean1.append(i)
clean1=' '.join(clean1)
clean1
```

```python
len(clean1)
```

#### frequency plot

```python
from nltk import FreqDist
```

```python
freq = FreqDist(word_tokenize(a))

freq.plot(20,cumulative=False);
```

```python
freq1 = FreqDist(word_tokenize(clean))

freq1.plot(20,cumulative=False);
```

```python
freq2 = FreqDist(word_tokenize(clean1))

freq2.plot(20,cumulative=False);
```

#### Using a function to plot

```python
def clean_text(text):
    token1 = word_tokenize(text)
    token2 = [word for word in token1 if word not in string.punctuation]
    token3 = [word.lower() for word in token2 if word.lower() not in swords]
    
    freq = FreqDist(token3)
    freq.plot(20)
    
#     return token3
```

```python
clean_text(a)
```
