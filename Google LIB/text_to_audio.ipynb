```python
from gtts import gTTS  #class
from playsound import playsound  # method
```

```python
text = """
फळा पुसायला भाऊंनी घेतलाय डस्टर … पोरी म्हणतात मलाच बनवा तुमची होम मिनिस्टर

डब्बे मे डब्बा, डब्बे मे दही.. बुलाती हे मगर जाणे का नाही..??

गावरान अंडी तळली तुपात…?
कायतरी जादू आहे भाऊच्या रुपात..?

भाऊंची एकच नजर आणि पोरींनी घेतला डोक्यावर पदर।।।।????

ती बोलली “मला तुझ्या हृदयात जागा दे ना.
भाऊ बोलले “मी आता अभ्यास करतोय तू उद्या ये ना ?

"""
```

```python
tts = gTTS(text = text ,lang = 'mr')
```

#### Save and Play the Audio 

```python
tts.save('speech.mp3')
import IPython.display as ipd
ipd.Audio(filename = 'speech.mp3')
```
