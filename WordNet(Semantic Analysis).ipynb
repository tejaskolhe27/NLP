#### Find the synonyms of all the adjectives in the sentence(except the same word)

```python
sent = '''The fiber is most often spun into yarn or thread and 
used to make a soft, rigid, and flexible textile.'''
```

```python
adj = [word[0] for word in pos_tag(word_tokenize(sent))
      if word[1].startswith("JJ")]
adj
```

```python
for word in adj:
    syns = set()
    for synset in wordnet.synsets(word):
        syns.add(synset.lemmas()[0].name())
    syns.discard(word)
    print("Synonyms of",word , 'are',syns)
```

# ANTONYMS

```python
ant = set()

for synset in wordnet.synsets("run"):
    for lemma in synset.lemmas():
        if lemma.antonyms():                     #When antonyms are present add to the set 
            ant.add(lemma.antonyms()[0].name())
print("Antonyms: ",ant)
```

```python
ant = set()

for synset in wordnet.synsets("soft"):
    for lemma in synset.lemmas():
        if lemma.antonyms():                     #When antonyms are present add to the set 
            ant.add(lemma.antonyms()[0].name())
print("Antonyms: ",ant)
```
