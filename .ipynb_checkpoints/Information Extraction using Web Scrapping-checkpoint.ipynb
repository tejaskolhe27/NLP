### Remove stopwords and print most frequent words

```python
from nltk import word_tokenize
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
import string
from nltk import FreqDist
```

```python
soup = BeautifulSoup(data,'html.parser')
text = soup.get_text()
```

```python
# Word tokenizing
token1 = word_tokenize(text)
```

```python
#Removing punctuations
token2 = [word for word in token1 if word not in string.punctuation]
```

```python
# Removing stopwords
swords=stopwords.words('english')
token3 = [word.lower() for word in token2 if word.lower() not in swords]
```

```python
# Removing stems/affixes
por_stem = PorterStemmer()
stem=[]

for word in token3:
    stem.append(por_stem.stem(word,to_lowercase=True))
```

```python
freq = FreqDist(stem)
freq.plot(3);
```

```python
freq.most_common()[0:3]
```

#### using new URL

```python
url=urllib.request.urlopen('https://en.wikipedia.org/wiki/Rahul_Dravid')
data=url.read()

ps=PorterStemmer()
soup=BeautifulSoup(data,'html.parser')
text=soup.get_text()

tokens=word_tokenize(text)
swords=stopwords.words('english')

tokens1=[x.lower() for x in tokens if x.isalpha() or x.isdigit()]
tokens2=[ps.stem(x) for x in tokens1 if x not in swords]
freq=FreqDist(tokens2)
freq.plot(10)
freq.most_common(3)
```

```python
from nltk import pos_tag
url=urllib.request.urlopen('https://en.wikipedia.org/wiki/Rahul_Dravid')
data=url.read()
```

```python
soup=BeautifulSoup(data,'html.parser')
text=soup.get_text()
```

```python
# Getting POS Tags
tokens=word_tokenize(text)

tags = pos_tag(tokens)
tags
```

```python
# Getting only verbs

count = 0
verbs = []

for t in tags:
    if t[1].startswith('VB'):
        count+=1
        verbs.append(t[0])
print("The count of verb is: ",count)
```

```python
verbs
```

```python
# Removing Punctuations
tokens1 = [word for word in verbs if word not in string.punctuation]
```

```python
#Removing stopwords and lemmatizing
from nltk import WordNetLemmatizer
swords = stopwords.words('english')
wsl = WordNetLemmatizer()

tokens2 = [wsl.lemmatize(x,'v') for x in tokens1 if x not in swords]
tokens2
```

```python
freq=FreqDist(tokens2)
freq.plot(10)
freq.most_common(3)
```
